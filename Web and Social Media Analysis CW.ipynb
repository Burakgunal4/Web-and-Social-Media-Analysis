```python
# Install NLTK if not already installed (run only once)
!pip install nltk

# Import necessary libraries
import nltk
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt

# Download tokenizer from NLTK (needed for some setups)
nltk.download('punkt')

# Step 1: Prepare the cleaned comments as a list of strings
text_data = df_clean['comment_clean'].tolist()

# Step 2: Create bigrams using CountVectorizer from sklearn
# ngram_range=(2, 2) means we only want 2-word combinations
vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
X = vectorizer.fit_transform(text_data)

# Step 3: Count the frequencies of each bigram
bigrams_freq = X.sum(axis=0)
bigrams = [(word, bigrams_freq[0, idx]) for word, idx in vectorizer.vocabulary_.items()]

# Step 4: Sort and take the top 15 most frequent bigrams
bigrams = sorted(bigrams, key=lambda x: x[1], reverse=True)[:15]

# Step 5: Convert the bigrams into a DataFrame for easy visualization
bigram_df = pd.DataFrame(bigrams, columns=['bigram', 'count'])

# Step 6: Plot a horizontal bar chart of the top 15 bigrams
plt.figure(figsize=(12, 6))
plt.barh(bigram_df['bigram'], bigram_df['count'], color='lightgreen')
plt.xlabel("Frequency")
plt.title("Top 15 Most Frequent Bigrams")
plt.gca().invert_yaxis()  # Show highest frequency at the top
plt.tight_layout()
plt.show()
```

```python
# Install nltk
!pip install nltk

# Import necessary libraries
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon (only required once)
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
vader = SentimentIntensityAnalyzer()
```

```python
# Apply VADER sentiment scoring to each comment
def get_sentiment_label(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Calculate sentiment scores and assign labels
df['compound_score'] = df['comment'].apply(lambda x: vader.polarity_scores(x)['compound'])
df['sentiment'] = df['compound_score'].apply(get_sentiment_label)

# Display a sample of results
df[['comment', 'compound_score', 'sentiment']].head(200)
```

```python
# Count the number of each sentiment label
sentiment_counts = df['sentiment'].value_counts()

# Convert counts to percentages
sentiment_percentages = sentiment_counts / sentiment_counts.sum() * 100

# Display the result
sentiment_percentages
```

```python
import matplotlib.pyplot as plt

# Pie chart of sentiment distribution
plt.figure(figsize=(6, 6))
plt.pie(sentiment_percentages, labels=sentiment_percentages.index, autopct='%1.1f%%', colors=['green', 'grey', 'red'])
plt.title("Sentiment Distribution (Pie Chart)")
plt.show()
```

```python
# Comment cleaning steps (if not already re-run)
import pandas as pd
import string
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# Assuming df is already defined (raw comments)
df['comment_clean'] = df['comment'].str.lower()
df['comment_clean'] = df['comment_clean'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df['comment_clean'] = df['comment_clean'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df_clean = df[df['comment_clean'].str.split().str.len() >= 3]
df_clean = df_clean[df_clean['comment_clean'].str.strip().astype(bool)]
df_clean.reset_index(drop=True, inplace=True)
```

```python
# Tokenize and clean each comment
texts = [
    [word for word in comment.split() if word not in stop_words]
    for comment in df_clean['comment_clean']
]
```

```python
# Create dictionary and corpus
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# Build the LDA model (you can adjust num_topics)
lda_model = gensim.models.ldamodel.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=5,  # Number of the main topics
    random_state=42,
    passes=10,
    per_word_topics=True
)
```

```python
# Display the top words in each topic
topics = lda_model.print_topics(num_words=5)
for idx, topic in topics:
    print(f"Topic {idx+1}: {topic}")
```

```python
# Display the top words in each topic
topics = lda_model.show_topics(num_topics=5, num_words=5, formatted=False)

for i, topic in topics:
    keywords = ", ".join([word for word, prob in topic])
    print(f"Topic {i+1}: {keywords}")
```

```python
import matplotlib.pyplot as plt

# Number of topics and words per topic
num_topics = 5
topn = 5

# Set up subplots
fig, axes = plt.subplots(nrows=1, ncols=num_topics, figsize=(18, 5), sharey=True)

# Loop through topics
for i in range(num_topics):
    topic = lda_model.show_topic(i, topn=topn)
    words = [word for word, prob in topic]
    weights = [prob for word, prob in topic]

    axes[i].bar(words, weights, color='skyblue')
    axes[i].set_title(f"Topic {i+1}")
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_ylim(0, max(weights)*1.2)  # uniform y-scale

fig.suptitle("Top 5 Keywords in Each Topic", fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.85)  # add spacing for suptitle
plt.show()
```
