```python
# Import Counter for word counting and pandas for tabular data
from collections import Counter
import pandas as pd

# Count word frequencies in the text
word_counts = Counter(text.split())

# Create a DataFrame of the top 15 most common words
freq_df = pd.DataFrame(word_counts.most_common(15), columns=['word', 'count'])

# Display the frequency table
freq_df
```

```python
# Plot a bar chart using the frequency table
plt.figure(figsize=(10, 6))
plt.bar(freq_df['word'], freq_df['count'], color='skyblue')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 15 Most Frequent Words")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Install NLTK if not already installed (run only once)
!pip install nltk

# Import necessary libraries
import nltk
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt

# Download tokenizer from NLTK (needed for some setups)
nltk.download('punkt')

# Step 1: Prepare the cleaned comments as a list of strings
text_data = df_clean['comment_clean'].tolist()

# Step 2: Create bigrams using CountVectorizer from sklearn
# ngram_range=(2, 2) means we only want 2-word combinations
vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
X = vectorizer.fit_transform(text_data)

# Step 3: Count the frequencies of each bigram
bigrams_freq = X.sum(axis=0)
bigrams = [(word, bigrams_freq[0, idx]) for word, idx in vectorizer.vocabulary_.items()]

# Step 4: Sort and take the top 15 most frequent bigrams
bigrams = sorted(bigrams, key=lambda x: x[1], reverse=True)[:15]

# Step 5: Convert the bigrams into a DataFrame for easy visualization
bigram_df = pd.DataFrame(bigrams, columns=['bigram', 'count'])

# Step 6: Plot a horizontal bar chart of the top 15 bigrams
plt.figure(figsize=(12, 6))
plt.barh(bigram_df['bigram'], bigram_df['count'], color='lightgreen')
plt.xlabel("Frequency")
plt.title("Top 15 Most Frequent Bigrams")
plt.gca().invert_yaxis()  # Show highest frequency at the top
plt.tight_layout()
plt.show()
```

```python
# REVISED API DATA RETENTION CODE (Time and user information with comments)

from googleapiclient.discovery import build
import pandas as pd

# Set the API key and video ID
api_key = "AIzaSyC0dslUAMHingkwPs7AEdRkiHV9SrFLzew"
video_id = "wgVOgGLtPtc"

# Initialize API client
youtube = build('youtube', 'v3', developerKey=api_key)

# Retrieve comments
comments = []
next_page_token = None

while len(comments) < 200:
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100,
        textFormat="plainText",
        pageToken=next_page_token
    )
    response = request.execute()

    for item in response["items"]:
        snippet = item["snippet"]["topLevelComment"]["snippet"]
        text = snippet["textDisplay"]
        likes = snippet.get("likeCount", 0)
        author = snippet.get("authorDisplayName", "Anonymous")
        published = snippet.get("publishedAt", None)

        comments.append({
            "comment": text,
            "likes": likes,
            "author": author,
            "publishedAt": published
        })

        if len(comments) >= 200:
            break

    next_page_token = response.get("nextPageToken")

# Create DataFrame
df = pd.DataFrame(comments)
print(df.head(200))


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Convert the 'publishedAt' column to datetime format
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

# Step 3: Extract only the date part (exclude time)
df['date_only'] = df['publishedAt'].dt.date

# Step 4: Count number of comments for each date
comments_by_day = df.groupby('date_only').size()

# Step 5: Plot a time-series line graph of daily comment counts
plt.figure(figsize=(10, 5))
comments_by_day.plot(kind='line', marker='o', color='purple')
plt.title("Number of Comments Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Comments")
plt.grid(True)
plt.tight_layout()
plt.show()
```
